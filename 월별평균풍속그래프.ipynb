```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 파일 경로
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시'에서 'year'와 'month'를 추출
weather['year'] = pd.DatetimeIndex(weather['일시']).year
weather['month'] = pd.DatetimeIndex(weather['일시']).month

# 월별 평균 풍속 계산
monthly_means = weather.groupby('month')['평균풍속'].mean()

# 월별 평균 풍속 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(monthly_means.index, monthly_means.values, color='blue')

# 그래프 라벨과 제목 설정
plt.xlabel('Month')
plt.ylabel('Average Wind Speed (m/s)')
plt.title('Monthly Average Wind Speed')

# X축에 월 이름을 표시 (1~12 대신 월 이름 사용)
plt.xticks(np.arange(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 그래프 출력
plt.show()
```
